```python
import pandas as pd

df = pd.read_json('../data/auto.json')
df.set_index('CarNumber',inplace=True)
df
```

```python
filtered_fines = df[df['Fines']>2100]
filtered_fines
```

```python
filtered_df = df[(df['Fines']>2100) & (df['Refund'] == 2)]
filtered_df
```

```python
filtered_models = df[df['Model'].isin(['Focus','Corolla'])]
filtered_models
```

```python
filtered_car = df[df.index.isin(['Y7689C197RUS','92928M178RUS','7788KT197RUS','H115YO163RUS','X758HY197RUS'])]
filtered_car
```

```python
median_fines_by_make = df.groupby('Make')['Fines'].median()
median_fines_by_make
```

```python
aggregated = df.groupby(['Make','Model'])['Fines'].count().reset_index(name='NumFines')
aggregated
```

```python
max_fines = df.groupby(['Make','Model'])['Fines'].max().reset_index(name='MaxFines')
max_fines
```

```python
min_fines = df.groupby(['Make','Model'])['Fines'].min().reset_index(name='MinFines')
min_fines
```

```python
std_fines = df.groupby(['Make','Model'])['Fines'].std().reset_index(name='Standard Deviation Fines')
std_fines
```

```python
car_fines = df.groupby('CarNumber')['Fines'].count().sort_values(ascending=False).reset_index(name='number of car fines')
car_fines
```

```python
fines_per_car = df.groupby(df.index).size()


max_fines = fines_per_car.max()
top_offenders = fines_per_car[fines_per_car == max_fines]
top_offenders
```

```python
top_1 = df.loc[df.index[0]]
top_1
```

```python
paid_CarNumber = df.groupby('CarNumber')['Fines'].sum().sort_values(ascending=False)
paid_CarNumber
```

```python
df.loc[paid_CarNumber.index[0]]
```
