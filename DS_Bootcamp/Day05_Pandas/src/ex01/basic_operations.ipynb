```python
import pandas as pd

df = pd.read_csv('../data/feed-views.log',sep='\t', names = ['datetime','user'])

df.head()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df.head()
```

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['minute'] = df['datetime'].dt.minute
df['second'] = df['datetime'].dt.second
df.head(10)
```

```python
bins = [0,4,7,11,17,20,24]
labels = ['night','early morning','morning','afternoon','early evening','evening']
df['daytime'] = pd.cut(df['hour'],bins = bins, labels = labels, right=False, include_lowest=True)
df.set_index("user",inplace=True)
df.tail()
```

```python
total_elements = df.count()
total_elements
```

```python
time_of_day_counts = df['daytime'].value_counts()
time_of_day_counts
```

```python
df_sorted = df.sort_values(by=['hour','minute','second'],ascending=True)
df_sorted
```

```python
min_hour = df['hour'].min()
max_hour = df['hour'].max()
mode_daytime = df['daytime'].mode()[0]
```

```python
max_hour_night = df[df['daytime'] == 'night']['hour'].max()
max_hour_night
min_hour_morning = df[df['daytime'] =='morning']['hour'].min()
```

```python
visitor_at_max_hour_night = df[(df['daytime'] == 'night') & (df['hour'] == max_hour_night)]
visitor_at_max_hour_night
```

```python
visitor_at_min_hour_morning = df[(df['daytime'] == 'morning') & (df['hour'] == min_hour_morning)]
visitor_at_min_hour_morning
```
